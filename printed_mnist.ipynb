```python
import numpy as np
import torch
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
import os
```

```python
labels = os.listdir('/content/gdrive/MyDrive/Sudoku/printed_mnist/assets')
print(labels)
```

```python
import PIL
X_2 = []
y_2 = []
for label in labels:
  images_path = os.path.join('/content/gdrive/MyDrive/Sudoku/printed_mnist/assets', label)
  images = os.listdir(images_path)
  for img in images:
        image = PIL.Image.open(os.path.join(images_path, img))
        image = image.convert("L")
        X_2.append(np.array(image).reshape(784))
        y_2.append(label)

X_2 = np.array(X_2)
y_2 = np.array(y_2,dtype = np.int)

X_2 = ((X_2-np.mean(X_2))/(np.std(X_2)))
X_2 = X_2.reshape(X_2.shape[0],1,28,28)
print(X_2[0],y_2[0])


  
```

```python
X_2 = torch.tensor(X_2)
y_2 = torch.tensor(y_2)
y_2 = y_2.type(torch.LongTensor)
```
